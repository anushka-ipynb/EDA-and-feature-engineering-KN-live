#### feature engineering

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.info()
```

```python
# -- another way of doing it --
# df["Date"]=df['Date_of_Journey'].apply(lambda x:x.split("/")[0])
# df["Month"]=df['Date_of_Journey'].apply(lambda x:x.split("/")[1])
# df["Year"]=df['Date_of_Journey'].apply(lambda x:x.split("/")[2])
```

```python
## converting into int now 
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df["Arrival_Time"]=df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
df['Arrival_Time']
```

```python
df.isnull().sum() #route and total stops have 1 null value. price having
#null values are okay, because that's from test data 
```

## we still have to handle arrival time -> it's in the hr and minutes format, we split it into two columns 

```python
df["arrival_hr"]=df['Arrival_Time'].apply(lambda x:x.split(":")[0])
df["arrival_mins"]=df['Arrival_Time'].apply(lambda x:x.split(":")[1])
```

```python
df['arrival_hr'] = df['arrival_hr'].astype(int)
df['arrival_mins'] = df['arrival_mins'].astype(int)
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df["Dep_Time"]=df['Dep_Time'].apply(lambda x:x.split(" ")[0])
df["dep_hr"]=df['Dep_Time'].apply(lambda x:x.split(":")[0])
df["dep_mins"]=df['Dep_Time'].apply(lambda x:x.split(":")[1])
df['dep_hr'] = df['dep_hr'].astype(int)
df['dep_mins'] = df['dep_mins'].astype(int)
df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df= df[~df['Total_Stops'].isnull()]
```

```python
df.isnull().sum()
```

```python
### mapping ranks to no of stops 
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df.drop(['Route'],axis=1,inplace=True)
```

```python
df['Duration'].value_counts()
```

```python
hrs = df['Duration'].str.split(' ').str[0]
mins = df['Duration'].str.split(' ').str[1]
```

```python
hrs = hrs.str[0]
mins = mins.str[0]
```

```python
hrs.fillna(0,inplace=True)
```

```python
mins.fillna(0,inplace=True)
```

```python
hrs = hrs.astype(int)
mins = mins.astype(int)
```

```python
df['Duration'] = hrs*60 + mins
```
