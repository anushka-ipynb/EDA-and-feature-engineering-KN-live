#### look at the number of columns that are getting generated if we take the normal one hot encoding approach - as the number of features increase, it increases the number of categorical one hot encoded features. this is called the curse of dimensionality

- so, how do we fix it?

##### in such situations, we take the 10(or 20, depending upon the data) categories, and make labels for them only. 

--- these are to undserstand how the function below will work---

```python
top_10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
for label in top_10:
    df[label] = np.where(df['X2']==label,1,0)

df.head(10)
```

***function begins here***

```python
def ohe(df,top_10,col):
    for label in top_10:
        df[col+"_"+label] = np.where(df[col]==label,1,0)
```

```python
for col in df.columns:
    top_10 = [x for x in df[col].value_counts().sort_values(ascending=False).head(10).index]
    ohe(df,top_10,col)
```
